social data about different parts of new yok city
create index to select out borough and the related data

```python
import pandas as pd
import numpy as np
```

```python
url = 'https://data.cityofnewyork.us/api/views/8ek7-jxw6/rows.csv?accessType=DOWNLOAD'

city_data = pd.read_csv(url)
```

```python
city_data.columns
```

Replace the * character with NaN

```python
city_data.replace('*', np.nan, inplace=True)
```

```python
city_data.head()
```

set the Dimension category Domain and Dimension as the index

```python
org_city_data = city_data.set_index(['Dimension Category', 'Domain', 'Dimension'])
org_city_data.head()
```

create new dataframe with only borough information

```python
borough_data = org_city_data.loc['Borough']
borough_data.head()
```

created dataframe with borough education data only

```python
borough_ed_data = borough_data.xs('Education', level='Domain')
borough_ed_data.head()
```

drop columns where all values are NaN
sort borough by four year high school gradutation percentage in 2015

Staten Island had the highest four year high school graduation percentagein 2015

```python
borough_ed_data.loc[borough_ed_data['Indicator'] == 'Four-Year High School Graduation Rate'].dropna(axis=1, how='all').sort_values('2015', ascending=False)
```

create a dataframe of crime by borough

```python
borough_crime_data = borough_data.xs('Personal & Community Safety', level='Domain')
borough_crime_data.head()
```

select the rows for violent crime per 1000 and drop columns that are all NaN
then sort from lowest to highest in 2015

as we can see the order of the boroughs is the same as the order of the boroughs who had the highest graduation rate
Staten Island has the lowest occurance of violent crime in 2015

```python
borough_crime_data.loc[borough_crime_data['Indicator'] == 'Violent Crime per 1000'].dropna(axis=1, how='all').sort_values('2015')
```
